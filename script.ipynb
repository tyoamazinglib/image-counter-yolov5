```python
model = torch.hub.load('ultralytics/yolov5', 'custom', path='yolov5/runs/train/exp67/weights/last.pt')
```

```python
model.eval()
```

```python
cap = cv2.VideoCapture('prochiz2.mp4')
while cap.isOpened():
    ret, frame = cap.read()
    
    # Make detections 
    results = model(frame)
    cv2.waitKey(50)
    cv2.imshow('Prochiz', np.squeeze(results.render()))
    
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
